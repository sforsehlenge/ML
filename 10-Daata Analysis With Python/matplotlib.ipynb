```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5,6]
y=[1,4,9,16,25,23]

## create  a linr plot
plt.plot(x,y)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('basic line plot')
plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[1,4,9,16,25,23]

## coustamize line plot
plt.plot(x,y,color='red', linestyle='-.', marker='o', linewidth=1, markersize=9) ## --, .,- :
plt.grid(1) ## or i can write True
```

```python
## multiple plots
## sample data
x= [1,2,3,4,5]
y1=[1,4,9,16,25]
y2=[1,2,3,4,5]

plt.figure(figsize=(9,5))
plt.subplot(2,2,1) ## means one row and two columb and make plot in 1st box
plt.plot(x,y1, color='green')
plt.title('plot 1')


plt.subplot(2,2,2) ## means one row and two columb and make plot in 1st box
plt.plot(x,y2, color='red')
plt.title('plot 2')


plt.subplot(2,2,3) ## means one row and two columb and make plot in 1st box
plt.plot(y1,y2, color='blue')
plt.title('plot 3')


plt.subplot(2,2,4) ## means one row and two columb and make plot in 1st box
plt.plot(y1,y2, color='black')
plt.title('plot 4')
```

```python
## bar plot

categories=['A','B','C','D','E']
value=[5,7,3,8,6]

plt.bar(categories, value, color='green')
```

```python
## create a scatter plot
#  sample data

x=[1,2,3,4,5]
y=[2,3,4,5,6]

plt.scatter(x,y,color='blue',marker='x')

```

```python
## pie chart

labels=['A','B','C','D']
sizes=[30,20,40,10]
colors=['gold','yellowgreen','lightcoral','lightskyblue']
explode=(0.2,0,0,0)

##create a pie chart
plt.pie(sizes,explode=explode,labels=labels,colors=colors,autopct="%1.1f%%",shadow=1)
```
