```python
import seaborn as sns
```

```python
## basic ploting with seaborn
tips=sns.load_dataset('tips')
tips
```

```python
## create a scater plot
import matplotlib.pyplot as plt

sns.scatterplot(x='total_bill', y= 'tip', data= tips)
plt.title("scatter plot of total biil vs tip")
plt.show()
```

```python
## line plot
sns.lineplot(x='size',y='total_bill',data=tips)
plt.title('line plot')
plt.show()
```

```python
## categoriacal plots
# barplot
sns.barplot(x='day', y='total_bill', data=tips)
plt.title('bar plot')
plt.show()
```

```python
# box plot
sns.boxplot(x='day',y='total_bill', data= tips)
plt.title('box plot')
plt.show()
```

```python
# violin plot

sns.violinplot(x='day', y='total_bill', data= tips)
plt.title('violin plot')
plt.show()
```

```python
#histogram
sns.histplot(tips['total_bill'],bins=10,kde=True) ## kde- kernal density estimator
```

```python
## kde plot
sns.kdeplot(tips['total_bill'],fill= True)
```

```python
# pair plot

sns.pairplot(tips) ## not take the categorial variable
```

```python
##heatmap

corr=tips[['total_bill','tip','size']].corr();
corr
```

```python
sns.heatmap(corr, annot=True, cmap='coolwarm')
```
